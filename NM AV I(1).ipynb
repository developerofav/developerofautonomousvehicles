```python
df.dropna(axis=0)
#dropna(axis=0): This function removes rows from the dataset where any element is missing (NaN/null). The axis=0 parameter specifies that rows with missing values should be dropped, helping to clean the dataset by eliminating incomplete records.
```

```python
df.dropna(axis=1)
#dropna(axis=1): This function removes columns from the dataset where any element is missing (NaN/null). The axis=1 parameter specifies that columns with missing values should be dropped, aiding in data cleaning by eliminating incomplete features or attributes.
```

```python
df.dropna(how='any')
#dropna(how='any'): This function removes rows or columns from the dataset where any element is missing (NaN/null). The parameter how='any' specifies that if any missing value is found in the row or column, it will be dropped, facilitating data cleaning by eliminating incomplete observations or features.
```

```python
df.dropna(how='all')
#dropna(how='all'): This function removes rows or columns from the dataset where all elements are missing (NaN/null). The parameter how='all' specifies that only rows or columns containing exclusively missing values will be dropped, useful for eliminating entirely empty records or features during data cleaning.
```

```python
df.dropna(thresh=50)
#dropna(thresh=50): This function removes rows from the dataset that have less than 50 non-null values. The thresh=50 parameter ensures that only rows with at least 50 non-null values are retained, which can be useful for filtering out incomplete records during data cleaning.
```

```python
#importing seaborn matplotlib and numpy for data visualization
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)
age_data=np.random.randint(20,60,size=100)
plt.hist(age_data,bins=10,color='black',edgecolor='white')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Histogram of Age Distribution')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
distance_traveled=np.random.uniform(0,100,size=50)
energy_consumption=np.random.uniform(5,20,size=50)
plt.scatter(distance_traveled,energy_consumption,color='green',alpha=0.7)
plt.xlabel('Distance traveled (km)'),plt.ylabel('Energy consumption(kWh)'),plt.title('Distance vs. Energy consumption Scatter plot')
plt.grid(True)
plt.show()
```
